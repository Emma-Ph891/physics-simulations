```python
import matplotlib.pyplot as plt

velocity_constant = 25
time_steps = [0, 2, 4, 6, 8, 10]

distance_steps = [velocity_constant * t for t in time_steps]
velocity_steps = [velocity_constant for t in time_steps]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.plot(time_steps, distance_steps, marker='o', color='blue')
ax1.set_title('Physics Simulation: Distance over Time')
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('Distance (meters)')
ax1.grid(True)

ax2.plot(time_steps, velocity_steps, marker='s', color='red', linestyle='--')
ax2.set_title('Physics Simulation: Velocity over Time')
ax2.set_xlabel('Time (seconds)')
ax2.set_ylabel('Velocity (m/s)')
ax2.set_ylim(0, velocity_constant + 10)
ax2.grid(True)

plt.tight_layout()
plt.savefig('physics_plots.png')
plt.show()

print(f"The object traveled {distance_steps[-1]} meters after {time_steps[-1]} seconds at a constant speed of {velocity_constant} m/s.")
```
