```python
import matplotlib.pyplot as plt
velocity = 25
time_steps = [0,2, 4, 6, 8, 10]
distance_steps = [velocity * t for t in time_steps]
print(f"the object traveled {distance_steps[-1]} meters after {time_steps[-1]} seconds.\n")
plt.plot(time_steps, distance_steps, marker='o', color='blue')
plt.title("physics simulation: Distance over time")
plt.xlabel("Time (seconds)")
plt.ylabel("Distance (meters)")
plt.grid(True)
plt.savefig('physics_plot.png')
plt.show
```
